**1.whta is the difference between a fnction and a method in python?**

In Python, both functions and methods are blocks of reusable code, but they differ mainly in how they are used.

**Function:**

A function is a standalone block of code that performs a specific task.

It is defined using the def keyword.

It can be called independently, outside of any object or class.

It takes input in the form of arguments and returns a value (optional).

```python
def greet(name): #function defination
    return "Hello " + name

print(greet("Mahesh"))  # Calling the function directly
```

**Method**:
A method is a function that is associated with an object.

It is also defined using the def keyword, but it is called on an instance of a class (i.e., an object).

It is part of a class and operates on data that belongs to the object of that class (often called self inside the method).

There are two types of methods: instance methods (which work on object data) and class/static methods (which can work without any object data).

in simple words

Method: A function that is defined inside a class and is called on an object (instance) of that class.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return "Hello " + self.name

p = Person("Mahesh")
print(p.greet())  # Calling the method on an instance of Person class
```

**2.Explain the concept of function arguments and parameters in Python?**

**Parameters :**Variables that define what kind of input the function will accept(during definition).

**Arguments**: The actual values passed to the function (during the function call).

**Types of arguments:**
positional,
keyword,
default,
arbitrary (*args and **kwargs).

**a. Positional Arguments :**
Positional arguments are the most common type. These arguments are passed to a function based on their position in the function call.

**b. Keyword Arguments :**
Keyword arguments are passed to a function using the parameter names explicitly. This allows you to pass arguments in any order, as long as the parameter name is used.

**c. Default Arguments :**
Default arguments allow you to define default values for parameters. If no argument is passed for a parameter, it will use the default value.

**d. Arbitrary (Variable-Length) Arguments**

Sometimes you may not know how many arguments will be passed to a function. You can use an asterisk (*) for positional arguments or two asterisks (**) for keyword arguments to accept a variable number of arguments.

  ***args:** Accepts a variable number of positional arguments.

  ****kwargs:** Accepts a variable number of keyword arguments.

**examples**

```python
def greet(name):  # 'name' is a parameter
    return f"Hello, {name}!"
print(greet("Mahesh"))  # 'Mahesh' is an argument
```

```python
#types of arguments
def add(a, b):
    return a + b

print("\n Positional arguments")
result = add(3, 5)  # 3 is passed to 'a', 5 is passed to 'b'
print(result)  # Output: 8

def add(a, b):
    return a + b

print("\n Keyword arguments")
result = add(b=5, a=3)  # The order doesn't matter because we use keywords
print(result)  # Output: 8

print("\n default arguments")

def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet())  # No argument passed, it will use the default 'Guest'
print(greet("Mahesh"))  # 'Mahesh' is passed, so it overrides the default

print("\n Arbitrary arguments")
def print_args(*args):
    for arg in args:
        print(arg)

print_args(1, "apple", True)

def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="Mahesh", age=25)
```

**5.What are iterators in Python and how do they differ from iterables?**

**Iterable:** An object capable of returning its members one at a time, like a list, tuple, or string. You can loop over it using a for loop.

**Examples:** lists, strings, and tuples.

**Iterator:** An object that keeps its state and produces the next value when requested using next(). It is created from an iterable using the iter() function and can be exhausted (meaning once values are consumed, they can’t be used again).

```python
print("Iterable \n")
my_list = [1, 2, 3]
for item in my_list:
    print(item)
print("\n Iterator \n")
my_iterator = iter(my_list)
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
```

**6. Explain the concept of generators in Python and how they are defined?**

**Generators** in Python are a special type of iterable that allow you to iterate over data one item at a time without storing the entire sequence in memory. They are defined using functions and the yield statement instead of return.

**Defined with yield:** Instead of returning a value and ending the function like return, yield pauses the function, keeping its state for the next iteration.

**Lazy evaluation:** Generators produce items only when requested (on demand), making them memory-efficient for large data sets.

Generators are great for handling large data streams where holding everything in memory isn't practical.

```python
def my_generator():
    for i in range(3):
        yield i

gen = my_generator()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
```

**7. What are the advantages of using generators over regular functions?**

Here are the key advantages of using generators over regular functions:

**Memory Efficiency:** Generators yield one item at a time, so they don't store the entire data in memory, which is ideal for large datasets.

**Lazy Evaluation:** Values are produced only when needed, improving performance and reducing computation time when not all values are required.

**State Retention:** Generators automatically retain their state between iterations, which eliminates the need for complex state-tracking logic in your code.

**Improved Performance:** Generators are faster in scenarios where only partial data is needed, as they avoid computing everything at once.

```python
# Regular function: returns a list of squares
def get_squares(n):
    return [i * i for i in range(n)]

# Generator: yields one square at a time
def square_generator(n):
    for i in range(n):
        yield i * i

# Regular function usage
squares_list = get_squares(5)
print("Using regular function:", squares_list)

# Generator usage
squares_gen = square_generator(5)
print("Using generator:")
for square in squares_gen:
    print(square)
```

**10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

**map()**

**Purpose:** Applies a given function to all items in an iterable (like a list) and returns a new iterable with the results.

**Return Type:** Returns a map object (which can be converted to a list).

**Use Case:** Ideal for transforming data (e.g., squaring numbers).

**filter()**

**Purpose:**Filters elements from an iterable based on a given function that returns either True or False.

**Return Type:** Returns a filter object (which can be converted to a list).

**Use Case:** Best for selecting elements that meet a certain condition (e.g., getting even numbers).

**reduce()**

**Purpose:** Applies a given function cumulatively to the items of an iterable, reducing it to a single value.

**Return Type:** Returns a single value.

**Use Case:** Useful for aggregating data (e.g., calculating the sum of a list of numbers).

```python
from functools import reduce

# Sample list
numbers = [1, 2, 3, 4, 5]

# Using map to square each number
squared = list(map(lambda x: x ** 2, numbers))  # Output: [1, 4, 9, 16, 25]

# Using filter to get even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))  # Output: [2, 4]

# Using reduce to get the sum of numbers
total = reduce(lambda x, y: x + y, numbers)  # Output: 15
```

**5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms**

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # Starting values for Fibonacci sequence
        self.count = 0  # To keep track of the current term

    def __iter__(self):
        return self  # Returns the iterator object

    def __next__(self):
        if self.count < self.n:
            fib = self.a  # The current Fibonacci number
            self.a, self.b = self.b, self.a + self.b  # Update to the next pair of numbers
            self.count += 1  # Move to the next term
            return fib
        else:
            raise StopIteration  # Stops the iteration when n terms are reached

# Example usage:
fib_iter = Fibonacci(10)  # Fibonacci sequence with 10 terms
for num in fib_iter:
    print(num)
```
